### Описание проекта

На основе данных по 500 абонентам компании «Мегалайн» за 2018 год необходимо проанализировать поведение абонентов и сделать вывод — какой тариф приносит больше денег. Это поможет коллегами из коммерческого департамента скорректировать рекламный бюджет.

#### Описание данных  
###### Таблица users (информация о пользователях): <a class="tocSkip">
user_id — уникальный идентификатор пользователя  
first_name — имя пользователя  
last_name — фамилия пользователя  
age — возраст пользователя (годы)  
reg_date — дата подключения тарифа (день, месяц, год)  
churn_date — дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)  
city — город проживания пользователя  
tariff — название тарифного плана  

###### Таблица calls (информация о звонках):<a class="tocSkip">
id — уникальный номер звонка  
call_date — дата звонка  
duration — длительность звонка в минутах  
user_id — идентификатор пользователя, сделавшего звонок  

###### Таблица messages (информация о сообщениях):<a class="tocSkip">
id — уникальный номер сообщения  
message_date — дата сообщения  
user_id — идентификатор пользователя, отправившего сообщение  

###### Таблица internet (информация об интернет-сессиях):<a class="tocSkip">
id — уникальный номер сессии  
mb_used — объём потраченного за сессию интернет-трафика (в мегабайтах)  
session_date — дата интернет-сессии  
user_id — идентификатор пользователя  

###### Таблица tariffs (информация о тарифах):<a class="tocSkip">
tariff_name — название тарифа  
rub_monthly_fee — ежемесячная абонентская плата в рублях  
minutes_included — количество минут разговора в месяц, включённых в абонентскую плату  
messages_included — количество сообщений в месяц, включённых в абонентскую плату  
mb_per_month_included — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)  
rub_per_minute — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)  
rub_per_message — стоимость отправки сообщения сверх тарифного пакета  
rub_per_gb — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

Так как таблиц довольно много, то напишем функцию, которая позволит нам выводить данные компактнее.  
На входе у неё будет список таблиц, список заголовков таблиц и общий заголовок(необязательный параметр).

```python
def display_side_by_side(dfs: list, captions: list, title=None):
    """Функция для компактного отображения таблиц.
    Аргументы:
    dfs -- список таблиц;
    captions -- список заголовков таблиц;
    title -- общий заголовок, необязательный параметр.
    """
    output = ""
    # проверка на длину списков
    if len(dfs) != len(captions):
        raise ValueError('Число таблиц и заголовков должно совпадать!')
    # объединяем аргументы в словарь
    combined = dict(zip(captions, dfs))
    if title is not None:
        print("\033[1m", title, "\033[0m")
    for caption, df in combined.items():
        # настраиваем параметры вывода
        output += df.style.set_table_attributes("style='display:inline'").set_caption(caption)._repr_html_()
        output += "\xa0\xa0\xa0"
    display(HTML(output))
```

Присвоим имена таблицам для удобства работы.

```python
users.name = 'users'
calls.name = 'calls'
messages.name = 'messages'
internet.name =  'internet'
tariffs.name = 'tariffs'
```

Посмотрим на наши таблицы.

```python
# объединим для того, чтобы сделать цикл по всем таблицам, кроме tariffs
data = [users, calls, messages, internet]
for df in data:
    print("\033[1m" "Таблица", df.name, "\033[0m")
    print("Общая информация, info()", '\n')
    display(df.info())
    display_side_by_side([df.head(15), df.sample(15, random_state=0), df.tail(15), df.describe()],
                         ['Первые 15 строк', 'Произвольные 15 строк', 'Последние 15 строк',
                          'Статистическая информация'])
    print("Доля пропущенных значений")
    display(df.isnull().mean().sort_values(ascending=False))
    print("Строк:", df.shape[0], '\n' "Дубликатов:", df.duplicated().sum())
    print()
```

```python
# таблица tariffs
print("\033[1m" "Таблица", tariffs.name, "\033[0m")
display(tariffs)
```

Напишем функцию, которая будет рассчитывать выручку.  
Вычтем бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; остаток умножим на значение из тарифного плана; прибавим абонентскую плату, соответствующую тарифному плану.

```python
def user_revenue_by_month(row):
    """Функция для расчёта выручки.
    Аргументы:
        row: строка таблицы.
    """
    # определяем тариф и ищем соответствующие значения лимитов в таблице с тарифами
    tariff = row['tariff']
    user_tariff = tariffs[tariffs['tariff_name'] == tariff].reset_index()
    calls_duration = row['calls_duration']
    sms_total = row['sms_total']
    mb_total = row['mb_total']
    monthly_fee = user_tariff.loc[0, 'rub_monthly_fee']
    minutes_included = user_tariff.loc[0, 'minutes_included']
    rub_per_minute = user_tariff.loc[0, 'rub_per_minute']
    messages_included = user_tariff.loc[0, 'messages_included']
    rub_per_message = user_tariff.loc[0, 'rub_per_message']
    mb_per_month_included = user_tariff.loc[0, 'mb_per_month_included']
    rub_per_gb = user_tariff.loc[0, 'rub_per_gb']

    revenue = 0
    # добавляем абонентскую плату
    revenue += monthly_fee

    # считаем перерасход за звонки
    if calls_duration > minutes_included:
        extra_minutes_cost = (calls_duration - minutes_included) * rub_per_minute
        revenue += extra_minutes_cost

    # за SMS-сообщения
    if sms_total > messages_included:
        extra_message_cost = (sms_total - messages_included) * rub_per_message
        revenue += extra_message_cost

    # за передачу данных    
    if mb_total > mb_per_month_included:
        extra_gb_used = np.ceil((mb_total - mb_per_month_included) / 1024)
        extra_traffic_cost = extra_gb_used * rub_per_gb
        revenue += extra_traffic_cost

    # возвращаем итоговое значение
    return revenue
```

```python
# применим функцию
user_stat_by_month['revenue'] = user_stat_by_month.apply(user_revenue_by_month, axis=1)
```

### Проверка гипотез

#### Cредняя выручка компании от абонентов тарифов «Ультра» и «Смарт» различается

Будем проверять две выборки на равенство их средних.  
Первая выборка - абоненты тарифа "Смарт", вторая - "Ультра".

Сформулируем гипотезы для статистического анализа:

H₀ - средняя выручка компании от абонентов тарифов «Ультра» и «Смарт» одинаковая.  
H₁ - средняя выручка компании от абонентов тарифов «Ультра» и «Смарт» различается. 

Уровень значимости зададим равным 5%

Сравним дисперсии, чтобы корректно выбрать значение equal_var. 

```python
print('Дисперсия выручки тарифа "Смарт"',
      int(user_stat_by_last_months.query('tariff == "smart"').loc[:, 'revenue'].describe()['std'] ** 2))
print('Дисперсия выручки тарифа "Ультра"',
      int(user_stat_by_last_months.query('tariff == "ultra"').loc[:, 'revenue'].describe()['std'] ** 2))
```

Дисперсии отличаются, оставим значение по умолчанию, т.е. equal_var = False.

```python
alpha = 0.05
results = st.ttest_ind(user_stat_by_last_months.query('tariff == "smart"').loc[:, 'revenue'],
                       user_stat_by_last_months.query('tariff == "ultra"').loc[:, 'revenue'])
print('P-value:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Для выбранного уровня значимости отвергаем нулевую гипотезу.  
Следовательно, средняя выручка от абонентов тарифов «Ультра» и «Смарт» различается.

#### Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

Первая генеральная совокупность - абоненты-москвичи, вторая - немосквичи.

Сформулируем гипотезы для статистического анализа:

H₀ - средняя выручка компании от абонентов в Москве равна выручке из других регионов  
H₁ - средняя выручка компании от абонентов в Москве НЕ равна выручке из других регионов

Уровень значимости зададим равным 5%

```python
# создадим список пользователей из Москвы и не из Москвы
moscow_users = users.query('city == "Москва"')['user_id'].value_counts().index.values
other_users = users.query('city != "Москва"')['user_id'].value_counts().index.values
```

Сравним дисперсии, чтобы корректно выбрать значение equal_var. 

```python
print('Дисперсия выручки в Москве',
      int(user_stat_by_last_months.query('user_id in @moscow_users').loc[:, 'revenue'].describe()['std'] ** 2))
print('Дисперсия выручки в регионах',
      int(user_stat_by_last_months.query('user_id in @other_users').loc[:, 'revenue'].describe()['std'] ** 2))
```

Дисперсия отличается, но не значительно, укажем equal_var = True.

```python
alpha = 0.05
results = st.ttest_ind(user_stat_by_last_months.query('user_id in @moscow_users').loc[:, 'revenue'],
                       user_stat_by_last_months.query('user_id in @other_users').loc[:, 'revenue'], equal_var=True)
print('P-value:', results.pvalue, )

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Для выбранного уровня значимости не получилось отвергнуть нулевую гипотезу.  
Следовательно, средняя выручка компании от абонентов в Москве равна выручке из других регионов
